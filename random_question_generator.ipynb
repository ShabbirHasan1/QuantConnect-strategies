```python
import random
from tkinter import *
from tkinter.ttk import *
```

```python
pytania_zbuw = []
card = open("question1.txt",'r')
for text in card.readlines():
    s = text.replace('\n','').replace('Å‚','l').replace('â€“','-').replace('Ã³','o').replace('Ä…','a').replace('Å›','s').replace('Ä‡','c').replace('Ä™','e').replace('Å¼','z').replace('Åº','z')
    pytania_zbuw.append(s)
del pytania_zbuw[0]

#########################################################

pytania_chirurgia = []
file = open("pytania_chirurgia.txt", 'r')
for q in file:
    q = q.replace('\n','').replace('Å‚','l').replace('â€“','-').replace('Ã³','o').replace('Ä…','a').replace('Å›','s').replace('Ä‡','c').replace('Ä™','e').replace('Å¼','z').replace('Åº','z').replace('ï»¿','')
    pytania_chirurgia.append(q)

#########################################################

pytania_kontaktologia = []
file = open("Pytania_kontaktologia.txt", 'r')
for q in file:
    q = q.replace('\n','').replace('Å‚','l').replace('â€“','-').replace('Ã³','o').replace('Ä…','a').replace('Å›','s').replace('Ä‡','c').replace('Ä™','e').replace('Å¼','z').replace('Åº','z').replace('ï»¿','')
    pytania_kontaktologia.append(q)

#########################################################

pytania_okulistyka = []
file = open("Pytania_okulistyka.txt", 'r')
for q in file:
    q = q.replace('\n','').replace('Å‚','l').replace('â€“','-').replace('Ã³','o').replace('Ä…','a').replace('Å›','s').replace('Ä‡','c').replace('Ä™','e').replace('Å¼','z').replace('Åº','z').replace('ï»¿','')
    pytania_okulistyka.append(q)


#########################################################

pytania_dziecieca = []
file = open("Pytania_dziecieca.txt", 'r')
for q in file:
    q = q.replace('\n','').replace('Å‚','l').replace('â€“','-').replace('Ã³','o').replace('Ä…','a').replace('Å›','s').replace('Ä‡','c').replace('Ä™','e').replace('Å¼','z').replace('Åº','z').replace('ï»¿','')
    pytania_dziecieca.append(q)

#########################################################

pytania_refrakcja = []
file = open("Pytania_refrakcja.txt", 'r')
for q in file:
    q = q.replace('\n','').replace('Å‚','l').replace('â€“','-').replace('Ã³','o').replace('Ä…','a').replace('Å›','s').replace('Ä‡','c').replace('Ä™','e').replace('Å¼','z').replace('Åº','z').replace('ï»¿','')
    pytania_refrakcja.append(q)

#########################################################

pytania_slabowidzenie = []
file = open("Pytania_slabowidzenie.txt", 'r')
for q in file:
    q = q.replace('\n','').replace('Å‚','l').replace('â€“','-').replace('Ã³','o').replace('Ä…','a').replace('Å›','s').replace('Ä‡','c').replace('Ä™','e').replace('Å¼','z').replace('Åº','z').replace('ï»¿','')
    pytania_slabowidzenie.append(q)
    del pytania_slabowidzenie[15:]

#########################################################


pytania_zaburzenia = []
file = open("Pytania_zaburzenia.txt", 'r')
for q in file:
    q = q.replace('\n','').replace('Å‚','l').replace('â€“','-').replace('Ã³','o').replace('Ä…','a').replace('Å›','s').replace('Ä‡','c').replace('Ä™','e').replace('Å¼','z').replace('Åº','z').replace('ï»¿','')
    pytania_zaburzenia.append(q)
del pytania_zaburzenia[::2]
del pytania_zaburzenia[15:]
```

```python
window = Tk()

def get_pytania_zbuw():
    question = random.choice(pytania_zbuw)
    w.config(text=question,aspect=300)
    
def get_pytania_chirurgia():
    question = random.choice(pytania_chirurgia)
    w.config(text=question,aspect=300)

def get_pytania_kontaktologia():
    question = random.choice(pytania_kontaktologia)
    w.config(text=question,aspect=300)

def get_pytania_okulistyka():
    question = random.choice(pytania_okulistyka)
    w.config(text=question,aspect=300)
    
def get_Pytania_dziecieca():
    question = random.choice(pytania_dziecieca)
    w.config(text=question,aspect=300)

def get_Pytania_refrakcja():
    question = random.choice(pytania_refrakcja)
    w.config(text=question,aspect=300)
    
def get_Pytania_slabowidzenie():
    question = random.choice(pytania_slabowidzenie)
    w.config(text=question,aspect=300)
    
def get_Pytania_zaburzenia():
    question = random.choice(pytania_zaburzenia)
    w.config(text=question,aspect=300)
    


w = Message(window, text="Hey Ola", width=450)
w.place(x=380, y=150, anchor="center")
w.configure(background='gray99')

style = Style() 
style.configure('W.TButton') 

b1 = Button(window, text = "Pytania ZBUW", command = get_pytania_zbuw,style = 'W.TButton')
b1.grid(row = 0, column = 0, padx=10, pady=10)
b1.config(width=20)

b2 = Button(window, text = "Pytania Chirurgia", command = get_pytania_chirurgia,style = 'W.TButton')
b2.grid(row = 1, column = 0, padx=10, pady=10)
b2.config(width=20)



b3 = Button(window, text = "Pytania Kontaktologia", command = get_pytania_kontaktologia,style = 'W.TButton')
b3.grid(row = 2, column = 0, padx=10, pady=10)
b3.config(width=20)

b4 = Button(window, text = "Pytania Okulistyka", command = get_pytania_okulistyka,style = 'W.TButton')
b4.grid(row = 3, column = 0, padx=10, pady=10)
b4.config(width=20)


b5 = Button(window, text = "Pytania Dziecieca", command = get_Pytania_dziecieca,style = 'W.TButton')
b5.grid(row = 4, column = 0, padx=10, pady=10)
b5.config(width=20)

b6 = Button(window, text = "Pytania Refrakcja", command = get_Pytania_refrakcja,style = 'W.TButton')
b6.grid(row = 5, column = 0, padx=10, pady=10)
b6.config(width=20)

b7 = Button(window, text = "Pytania Slabowidzienie", command = get_Pytania_slabowidzenie,style = 'W.TButton')
b7.grid(row = 6, column = 0, padx=10, pady=10)
b7.config(width=20)

b8 = Button(window, text = "Pytania Zaburzenia", command = get_Pytania_zaburzenia,style = 'W.TButton')
b8.grid(row = 7, column = 0, padx=10, pady=10)
b8.config(width=20)




window.geometry("700x360")
window.configure(background='gray99')
window.mainloop()
```
